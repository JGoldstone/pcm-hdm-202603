# CIE 1931 2° Standard Observer (Plotly)

Interactive plot of the CIE 1931 2° colour matching functions.
Hover any point to read wavelength (nm) and function value.


```python
import numpy as np
import plotly.graph_objects as go
from colour.colorimetry import MSDS_CMFS

cmfs = MSDS_CMFS['CIE 1931 2 Degree Standard Observer']
wavelengths = np.asarray(cmfs.wavelengths, dtype=np.float64)
x_bar = np.asarray(cmfs.values[..., 0], dtype=np.float64)
y_bar = np.asarray(cmfs.values[..., 1], dtype=np.float64)
z_bar = np.asarray(cmfs.values[..., 2], dtype=np.float64)

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=wavelengths, y=x_bar, mode='lines', name='x̄(λ)',
    line={'color': '#d62728', 'width': 2},
    hovertemplate='x̄(λ)<br>λ=%{x:.0f} nm<br>value=%{y:.6f}<extra></extra>'
))
fig.add_trace(go.Scatter(
    x=wavelengths, y=y_bar, mode='lines', name='ȳ(λ)',
    line={'color': '#2ca02c', 'width': 2},
    hovertemplate='ȳ(λ)<br>λ=%{x:.0f} nm<br>value=%{y:.6f}<extra></extra>'
))
fig.add_trace(go.Scatter(
    x=wavelengths, y=z_bar, mode='lines', name='z̄(λ)',
    line={'color': '#1f77b4', 'width': 2},
    hovertemplate='z̄(λ)<br>λ=%{x:.0f} nm<br>value=%{y:.6f}<extra></extra>'
))

fig.update_layout(
    title='CIE 1931 2° Standard Observer Colour Matching Functions',
    template='plotly_white',
    xaxis={'title': 'Wavelength (nm)'},
    yaxis={'title': 'Value'},
    hovermode='x unified',
    legend={'orientation': 'h', 'yanchor': 'bottom', 'y': 1.02, 'xanchor': 'left', 'x': 0},
    height=540,
)

fig.show()
```
